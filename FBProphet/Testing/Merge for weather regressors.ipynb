## Merge for weather regressors

```python
import pandas as pd
```

```python
df_train = pd.read_csv('2019_prophet.csv')
```

```python
df_train= df_train.rename(columns={'Date': 'ds', 'Total': 'y'})
df_train.tail()
```

```python
df_test = pd.read_csv('2020_prophet.csv')
```

```python
df_test = df_test.rename(columns={'Date': 'ds', 'Total': 'y'})
df_test.head()
```

```python
merge1 = pd.concat((df_train, df_test))
```

```python
merge1 = merge1.reset_index(drop=True)
```

```python
merge1
```

```python
temp = pd.read_csv('weather_prophet.csv')
```

```python
temp.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
temp.rename(columns={'date':'ds'}, inplace=True)
```

```python
temp
```

```python
merge2 = pd.merge(merge1,temp, on='ds')
```

```python
merge2
```

```python
merge2.to_csv('AllDataTemp.csv')
```

```python
merge2.drop(columns=['temp','rain'], inplace=True)
```

```python
merge2.rename(columns={'ds':'Date', 'y':'Total'}, inplace=True)
```
